```python
import sqlite3

conn = sqlite3.connect('mock_data.db')

print ("Initiated successfully")
```

```python
c = conn.cursor()
```

```python
c.execute("""CREATE TABLE "MOCK_DATABASE"(
    "Mk" TEXT ,
    "Category Type Approved" TEXT ,
    "Category Registered" TEXT ,
    "Mass (kg)" INT ,
    "CO2 (g/km)" INT ,
    "W (mm)" INT ,
    "Steering Axle (mm)" INT ,
    "Other Axle (mm)" INT ,
    "Ft" TEXT ,
    "Fm" TEXT ,
    "Engine Capacity (cm3)" INT ,
    "Engine Power (KW)" INT)""")
conn.commit()
print("Table was Created")
```

```python
file = "Resources/mock_data_for_practice.csv"
```

```python
import pandas as pd
df = pd.read_csv(file, index_col=0)
```

```python
df.head()
```

```python
df.to_sql("MOCK_DATABASE", conn, if_exists = 'append', index=0)
```

```python
conn.execute("SELECT * FROM MOCK_DATABASE").fetchall()
```
