```python
import pandas as pd
import os
```

```python
file = os.path.join("..","Resource","clean_raw_data.csv")
df = pd.read_csv(file)
df.head()
```

```python
trans_df = df[['ID','Mk', 'Ct','Cr', 'm (kg)','Enedc (g/km)','W (mm)','At1 (mm)','At2 (mm)','Ft','Fm','ec (cm3)','ep (KW)']]
trans_df.head(10)
```

```python
trans_df.rename({'Ct': 'Category Type Approved','Cr':'Category Registered', 'm (kg)':'Mass (kg)','Enedc (g/km)':'CO2 (g/km)','At1 (mm)':'Steering Axle (mm)','At2 (mm)':'Other Axle (mm)','ec (cm3)':'Engine Capacity (cm3)','ep (KW)':'Engine Power (KW)'}, axis=1, inplace=True)
trans_df.head(20)
```

```python
car_size_df = trans_df[['ID','Mass (kg)','CO2 (g/km)','W (mm)','Steering Axle (mm)','Other Axle (mm)','Engine Capacity (cm3)','Engine Power (KW)']]
car_size_df.head()
```

```python
car_size = car_size_df.set_index('ID')
car_size.head()
```

```python
car_size.to_csv(os.path.join("..","Resource","car_size.csv"))
```

```python
car_category_df = trans_df[['ID','Mk', 'Category Type Approved','Category Registered','Ft','Fm']]
car_category_df.head()
```

```python
car_category = car_category_df.set_index('ID')
car_category.head()
```

```python
car_category.to_csv(os.path.join("..","Resource","car_category.csv"))
```
